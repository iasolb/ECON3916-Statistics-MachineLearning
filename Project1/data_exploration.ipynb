```python
import pandas as pd

from data_handling import MrozHandler
fp = "data/Mroz.csv"
Mroz = MrozHandler(fp)

print([col for col in Mroz.full.columns])
```

## Summary Statistics:

Mean, Standard Deviation for full dataset, working subset, non-working, subset

```python
columns = ["lwage", "exper", "educ", "kidslt6", "nwifeinc", "agew", "motheduc", "fatheduc"]

working = Mroz.working
non_working = Mroz.full[~Mroz.full.index.isin(working.index)]

summary = pd.DataFrame({
    "Mean (W)":  working[columns].mean(),
    "SD (W)":    working[columns].std(),
    "Mean (NW)": non_working[columns].mean(),
    "SD (NW)":   non_working[columns].std(),
    "Mean (All)": Mroz.full[columns].mean(),
    "SD (All)":   Mroz.full[columns].std(),
}).round(2)

# lwage not observed for non-working
summary.loc["lwage", ["Mean (NW)", "SD (NW)"]] = "—"

print(summary.to_string())

```
