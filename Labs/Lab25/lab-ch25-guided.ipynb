# Part 6: Reflection

## Q1: What worked well with AI-assisted development?

AI-assisted development dramatically accelerated the scaffolding phase — generating a full Next.js component tree, Tailwind layout, and placeholder content in minutes rather than hours. Code quality for boilerplate tasks (responsive grid, color theming, repeated card components) was consistently solid and production-ready. Claude handled repetitive structural patterns well, such as mapping over project data to render uniform cards, which would have been tedious to write by hand. This aligns with Chapter 25's framing of AI as a productivity multiplier for well-defined, pattern-heavy subtasks where the solution space is narrow.

## Q2: What did you have to fix manually?

Several layout details required manual correction: the mobile navigation collapsed incorrectly, and section spacing assumed a fixed viewport that broke on smaller screens. Claude occasionally hallucinated component props or Tailwind utility classes that either don't exist or behaved differently than described, requiring verification against actual docs. The hero section's typography hierarchy needed manual tuning — the AI defaulted to generic sizing that didn't reflect the visual weight I wanted. These fixes reflect Chapter 25's point that AI outputs are probabilistic approximations, not specifications, and human judgment remains essential for aesthetic and UX decisions.

## Q3: How does AI-assisted development compare to writing code entirely yourself?

The HADI loop played out clearly: I hypothesized that a prompt describing a "minimal economist portfolio with dark theme" would yield a usable scaffold (Hypothesis), submitted it (Action), reviewed the output for structural correctness and visual fit (Data), and refined my next prompt based on what was off (Insight). This cycle ran much faster than solo coding because the AI collapsed the implementation step, letting me spend cognitive load on evaluation instead of syntax. However, domain knowledge gaps were real — the AI didn't know my specific project requirements, aesthetic preferences, or the economics framing I wanted to convey, so every iteration required me to inject that context deliberately. Writing entirely solo is slower but produces output more tightly coupled to intent from the start.

## Q4: Prompt engineering lessons

Good prompts are specific about constraints (framework, style system, target audience) and desired output format, rather than open-ended descriptions of goals. Bad prompts are vague or assume the AI shares implicit context — "make it look good" produces generic results because "good" is undefined. One concrete improvement: I initially prompted "build a portfolio page" and got a generic template; I revised to "build a single-page Next.js 14 portfolio for an economics researcher, using Tailwind, dark background, sections for About / Projects / Skills / Contact, with a professional sans-serif type scale" — the output was immediately closer to usable, requiring far fewer follow-up corrections.

## Q5: Would a portfolio site influence hiring a junior data scientist?

Yes, meaningfully so — a portfolio site signals effort, aesthetic judgment, and communication skill that a resume cannot. Per labor economics signaling theory (Spence), credentials on a resume are cheap to claim but a live, polished project requires actual investment of time and taste, making it a credible signal of conscientiousness and technical follow-through. A junior candidate whose portfolio clearly explains *what* they built, *why* it matters, and *what they learned* demonstrates the ability to frame technical work for a non-technical audience — a skill critical in applied economics and data science roles. Conversely, a poorly organized or broken portfolio can *negatively* signal attention to detail, so the signal cuts both ways.
